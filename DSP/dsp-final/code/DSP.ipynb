```python
# import matplotlib 相關套件

import matplotlib.pyplot as plt

bits = [10, 9, 8, 7, 6, 5, 4, 3]

t1 = [312726, 231819, 227560, 218722, 214666, 213978, 213280, 212974]
t2 = [349487, 267600, 264885, 258902, 256183, 254166, 253364, 253078]
t3 = [329182, 247982, 240985, 231373, 225771, 225934, 223457, 223633]
t4 = [467713, 355130, 361756, 346586, 350569, 346771, 344646, 344657]
t5 = [440491, 330324, 331842, 323349, 321964, 318889, 317772, 318124]

# 設定圖片大小為長15、寬10

plt.figure( figsize=(15,10), dpi=100, linewidth = 2)
plt.plot( bits, t1,'s-',color = 'b', label="T1")
plt.plot( bits, t2,'s-',color = 'g', label="T2")
plt.plot( bits, t3,'s-',color = 'r', label="T3")
plt.plot( bits, t4,'s-',color = 'c', label="T4")
plt.plot( bits, t5,'s-',color = 'm', label="T5")

plt.xticks(fontsize=20)

plt.yticks(fontsize=20)

# 標示x軸(labelpad代表與圖片的距離)

plt.xlabel("Bits", fontsize=30, labelpad = 15)

# 標示y軸(labelpad代表與圖片的距離)

plt.ylabel("Bytes", fontsize=30, labelpad = 20)

# 顯示出線條標記位置

plt.legend(loc = "best", fontsize=20)

# 畫出圖片

plt.show()
```
